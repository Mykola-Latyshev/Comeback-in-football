```python
import pandas as pd
import numpy as np
```

```python
#load data
df_en_2020 = pd.read_csv('data\df_en_2020.csv')
df_en_2020_odd_start = pd.read_csv('data\England00_2020.csv')

df_en_2019 = pd.read_csv('data\df_en_2019.csv')
df_en_2019_odd_start = pd.read_csv('data\England00_2019.csv')

df_en_2018 = pd.read_csv('data\df_en_2018.csv')
df_en_2018_odd_start = pd.read_csv('data\England00_2018.csv')
```

```python
# slice only important columns
need_columns_2020 = ['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']
df_en_2020_odd_start = pd.DataFrame(df_en_2020_odd_start, columns=need_columns_2020)
df_en_2020_odd_start.rename(columns = {'B365H':'HW', 'B365D':'D', 'B365A':'AW','B365>2.5':'>2.5', 'B365<2.5':'<2.5'}, inplace = True)

need_columns_2019 = ['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'B365H', 'B365D', 'B365A', 'B365>2.5', 'B365<2.5']
df_en_2019_odd_start = pd.DataFrame(df_en_2019_odd_start, columns=need_columns_2019)
df_en_2019_odd_start.rename(columns = {'B365H':'HW', 'B365D':'D', 'B365A':'AW','B365>2.5':'>2.5', 'B365<2.5':'<2.5'}, inplace = True)

need_columns_2018 = ['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'B365H', 'B365D', 'B365A', 'BbAv>2.5', 'BbAv<2.5']
df_en_2018_odd_start = pd.DataFrame(df_en_2018_odd_start, columns=need_columns_2018)
df_en_2018_odd_start.rename(columns = {'B365H':'HW', 'B365D':'D', 'B365A':'AW','BbAv>2.5':'>2.5', 'BbAv<2.5':'<2.5'}, inplace = True)
```

```python
# check the teams name from different database
df01 = df_en_2020_odd_start['HomeTeam'].unique()
df02 = df_en_2020['name_team_home'].unique()
set(df01) - set(df02)
```

```python
def rename_en_team_names(df_odd):
    df_odd.loc[df_odd.HomeTeam == 'Man United', 'HomeTeam'] = 'Manchester United'
    df_odd.loc[df_odd.AwayTeam == 'Man United', 'AwayTeam'] = 'Manchester United'

    df_odd.loc[df_odd.HomeTeam == 'Man City', 'HomeTeam'] = 'Manchester City'
    df_odd.loc[df_odd.AwayTeam == 'Man City', 'AwayTeam'] = 'Manchester City'

    df_odd.loc[df_odd.HomeTeam == 'Sheffield United', 'HomeTeam'] = 'Sheffield Utd'
    df_odd.loc[df_odd.AwayTeam == 'Sheffield United', 'AwayTeam'] = 'Sheffield Utd'
    return(df_odd)
```

```python
df_en_2020_odd_start = rename_en_team_names(df_en_2020_odd_start)
df_en_2019_odd_start = rename_en_team_names(df_en_2019_odd_start)
df_en_2018_odd_start = rename_en_team_names(df_en_2018_odd_start)
```

```python
df_en_2018_odd_start.head()
```

```python
def join_df(df, df_odd):
    df_final = pd.merge(df, df_odd, how = 'left', 
                      left_on=['name_team_home','name_team_away'], right_on = ['HomeTeam','AwayTeam'])
    return(df_final)
```

```python
df_en_2020_final = join_df(df_en_2020, df_en_2020_odd_start)
df_en_2019_final = join_df(df_en_2019, df_en_2019_odd_start)
df_en_2018_final = join_df(df_en_2018, df_en_2018_odd_start)

df_en_final = pd.concat([df_en_2018_final, df_en_2019_final, df_en_2020_final]).reset_index()
```

```python
df_en_final.head()
```

```python
# add column 'comeback' 0/1 - comeback during last 15+ minutes (from 75 to end)

df_en_final['match_dif'] = df_en_final['0-75_dif'] + df_en_final['76-90+_dif']
for i in range(0, len(df_en_final)):
    if (df_en_final.loc[i, '0-75_dif'] < 0):
        if (df_en_final.loc[i, 'match_dif'] < 0):
            df_en_final.loc[i, 'comeback'] = 0
        else:
            df_en_final.loc[i, 'comeback'] = 1
     
    if (df_en_final.loc[i, '0-75_dif'] == 0):
        if (df_en_final.loc[i, 'match_dif'] == 0):
            df_en_final.loc[i, 'comeback'] = 0
        else:
            df_en_final.loc[i, 'comeback'] = 1
            
    if (df_en_final.loc[i, '0-75_dif'] > 0):
        if (df_en_final.loc[i, 'match_dif'] > 0):
            df_en_final.loc[i, 'comeback'] = 0
        else:
            df_en_final.loc[i, 'comeback'] = 1
```

```python
# add difference beetwen goals homa and away teams (class less= -2, -1, 0, 1, more= 2)
for i in range(0, len(df_en_final)):
    if (df_en_final.loc[i, '0-75_dif'] <= -2):
        df_en_final.loc[i, 'dif_class'] = -2
    if (df_en_final.loc[i, '0-75_dif'] == -1):
        df_en_final.loc[i, 'dif_class'] = -1
    if (df_en_final.loc[i, '0-75_dif'] == 0):
        df_en_final.loc[i, 'dif_class'] = 0
    if (df_en_final.loc[i, '0-75_dif'] == 1):
        df_en_final.loc[i, 'dif_class'] = 1    
    if (df_en_final.loc[i, '0-75_dif'] >= 2):
        df_en_final.loc[i, 'dif_class'] = 2  
```

```python
df_en_final = df_en_final.drop(['index'], axis=1)
```

```python
# add columns with probability of win
df_en_final['per_coef_win'] = 1/df_en_final['HW'] + 1/df_en_final['D'] + 1/df_en_final['AW']

df_en_final['prob_HW'] = (1/df_en_final['HW'])/(df_en_final['per_coef_win'])
df_en_final['prob_D'] = (1/df_en_final['D'])/(df_en_final['per_coef_win'])
df_en_final['prob_AW'] = (1/df_en_final['AW'])/(df_en_final['per_coef_win'])
```

```python
# add columns with probability of total
df_en_final['per_coef_total'] = 1/df_en_final['>2.5'] + 1/df_en_final['<2.5']

df_en_final['prob_25more'] = (1/df_en_final['>2.5'])/(df_en_final['per_coef_total'])
df_en_final['prob_25less'] = (1/df_en_final['<2.5'])/(df_en_final['per_coef_total'])

df_en_final = df_en_final.drop(['per_coef_win', 'per_coef_total'], axis=1)
```

```python
df_en_final['id_country'] = 0
df_en_final['country'] = 'en'
```
